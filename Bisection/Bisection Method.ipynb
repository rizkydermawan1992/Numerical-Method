# Bisection Method

    Prosedur Pencarian Akar Fungsi

1.  Tentukan fungsi f(x), interval [a,b], nilai 
     toleransi (ε), dan iterasi maksimum (imax)
2.  Hitung masing-masing nilai f(a) dan f(b)
3.  Hitung nilai f(a).f(b) lalu cek hasilnya dimana :
      Jika f(a).f(b) < 0 maka lanjutkan proses
      Jika f(a).f(b) >= 0 maka hentikan proses 
4.  Hitung nilai tengah c = (a+b)/2 dan f(c)
5.  Hitung nilai f(c).f(a) lalu cek hasilnya dimana :
     Jika f(c).f(a) < 0 maka b diganti c dan f(b) diganti f(c)
     Jika f(c).f(a) > 0 maka a diganti c dan f(a) diganti f(c)
     Jika f(c).f(a) = 0 maka c adalah akar fungsi
6.  Hitung nilai error (|b-a|)
7.  Pengecekan akhir iterasi dimana :
     Jika |b-a| <=  ε maka hentikan iterasi (Akar Fungsi = c)
     Jika |b-a| > ε maka lanjutkan iterasi (ulangi
     Langkah 4)


```python
# Membuat input user
fungsi = input("Masukkan Fungsi f(x) = ")
a      = float(input("Masukkan Nilai a = "))
b      = float(input("Masukkan Nilai b = "))
eps    = float(input("Masukkan Nilai eps = "))
imax   = int(input("Masukan Nilai i max = "))

x1 = a
x2 = b

# Membuat function f(x)
def f(x):
    fx = eval(fungsi)
    return fx

# Menghitung nilai f(a) dan f(b)
fa = f(a)
fb = f(b)

# Mengecek hasil kali f(a) dan f(b), apakah bernilai positif atau negatif
if fa*fb >= 0:
    print("Tentukan nilai a dan b yang baru")
    exit()

# Membuat wadah untuk menampung data tiap iterasi
list_i = []
list_a = []
list_b = []
list_c = []
list_fa = []
list_fb = []
list_fc = []
list_error = []


# Melakukan iterasi
for i in range(1,imax):
    # Menghitung nilai tengah c dan f(c)
    c = (a+b)/2
    fc = f(c)
    # Menentukan nilai a dan b untuk iterasi selanjutnya
    if fc*fa < 0:
        b = c
        fb = fc
    elif fc*fa > 0:
        a = c
        fa = fc
    else: #fc*fa == 0
        break
    
    # Menghitung nilai error
    error = abs(b-a)
    
    # Mengisi data tiap iterasi ke dalam wadah
    list_i.append(i)
    list_a.append(a)
    list_b.append(b)
    list_c.append(c)
    list_fa.append(fa)
    list_fb.append(fb)
    list_fc.append(fc)
    list_error.append(error)
    
    # Mencetak data tiap iterasi
    print("i = ", i)
    print("a = %.5f"%a)
    print("b = %.5f"%b)
    print("c = %.5f"%c)
    print("f(a) = %.5f"%fa)
    print("f(b) = %.5f"%fb)
    print("f(c) = %.5f"%fc)
    print("|b-a| = %.5f"%error)
    print("--------------------------")
    
    # Mengecek kapan iterasi dihentikan
    if error <= eps:
        break


# Mencetak nilai akar fungsi
print("Akar Fungsi f(x) = ", c)
print("Jumlah Iterasi = ", i)
```

```python
# Menampilkan data dalam bentuk tabel
import pandas as pd
data ={
        "i" : list_i, 
        "a" : list_a,
        "b" : list_b,
        "c" : list_c,
        "fa" : list_fa,
        "fb" : list_fb,
        "fc" : list_fc,
        "|b-a|" : list_error
      }

tabel = pd.DataFrame(data)
tabel
```
